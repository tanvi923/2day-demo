## **Load Data**


```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv')
df
```

### data seperation as x n y

```python
y = df['logS']
y
```

```python
x = df.drop('logS', axis=1)
x
```

#### split the data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=100)
```

```python
X_train
```

```python
X_test
```

## **model building**

### linear regression

**training the model**

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, Y_train)
```

**applying the model to make predictions**

```python
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

**model performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(Y_train, y_lr_train_pred)
lr_train_r2 = r2_score(Y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(Y_test, y_lr_test_pred)
lr_test_r2 = r2_score(Y_test, y_lr_test_pred)
```

```python
print('LR MSE (Train): ', lr_train_mse)
print('LR R2 (Train): ', lr_train_r2)
print('LR MSE (Test): ', lr_test_mse)
print('LR R2 (Test): ', lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear Regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns=['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
lr_results
```

```python
lr_results
```

##**random forest**

##training the model

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(X_train, Y_train)
```

## **model performance**

### **applying the model to predict**

```python
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_rf_train_mse = mean_squared_error(Y_train, y_rf_train_pred)
y_rf_train_r2 = r2_score(Y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(Y_test, y_rf_test_pred)
rf_test_r2 = r2_score(Y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random Forest', y_rf_train_mse, y_rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns=['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
rf_results
```

## **model comparision**

```python
df_models = pd.concat([rf_results, lr_results], axis=0)
df_models
```

```python
df_models.reset_index(drop=True)
```

## **data visualization of pred result**

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(5,5))
plt.scatter(x=Y_train, y=y_lr_train_pred, c="#7CAE00", alpha=0.3)

z = np.polyfit(Y_train, y_lr_train_pred, 1)
p = np.poly1d(z)

plt.plot(y_lr_train_pred, p(y_lr_train_pred))
plt.ylabel('Predict LogS')
plt.xlabel('Experimental LogS')
plt.title('Linear Regression Model')
```
